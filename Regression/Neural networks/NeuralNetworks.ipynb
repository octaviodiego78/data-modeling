**¿Qué es una neurona?**

Es un nombre elegante que le pusieron para referirse a una función en los años de 1940-1950. Se creeía que los nodos eran como neuronas del cerebro.  

**¿Qué es una función?**

Es un nombre elegante que pusieron para referirse a algo que toma cosas de entrada, aplica alguna lógica y saca un resultado. 

**¿Qué es una red neuronal?**

Es la interconexión de múltiples neuronas (funciones) para resolver un problema complejo que una sola neurona no podría. Dentro de los componentes existen *nodos* y *conexiones*

**Pasos que hace la red neuronal:**

1. Combinación de las variables de entrada (X). Cada señal recibida tiene una constante w (peso/weight) que permite manipular la importancia de cada variable de entrada. 
2. Neurona de respuesta. Considerando todas las variables de entrada, la neurona genera una señal de respuesta a través de una función. 

**Tipos de capas en la red neuronal**

- Capa de entrada: capa que toma los datos de entrada
- Capas ocultas: toman las entradas de otra capa y pasan a salida hacia otra capa (son intermediarias entre la entrada y la salida)
- Capas de salida: capa que hace la predicción

**Funciones de activación**

El comportamiento de las neuronas de las redes neuronales es basado en la función en la que la neurona se especializó. Una neurona puede cambiar su respuesta dependiendo de la función de activación $\varphi(*)$ que sea seleccionada

Funciones comunes:
1. Lineal: $\varphi(\nu)=\nu$
2. Sigmoidal: $\varphi(\nu)=\frac{e^{\nu}}{1+e^{\nu}}$ 
3. Tanh: $\varphi(\nu)=\frac{e^{\nu}-e^{-\nu}}{e^{\nu}+e^{-\nu}}=tanh(\nu)$
4. Softplus: $\varphi(\nu)=log(1+e^{\nu})$
5. RELU: $\varphi(\nu)=max(0,\nu)$

Las funciones de activación suelen ser diferenciables, lo que significa que la derivada de primer orden se puede calcular para un valor de entrada dado. Esto es necesario dado que las redes neuronales generalmente se entrenan utilizando el algoritmo de backpropagation que requiere la derivada del error de predicción para actualizar los pesos del modelo.

**¿Cómo elegir la función de activación para las capas ocultas?**

- Lineal: se recomienda para cuando el problema es sencillo
- Sigmoidal: típicamente para problemas de clasificación. Sufre del problema de "Gradiente que se desvanece", los gradientes de la salida de la red con respecto a los parámetros se vuelven tan pequeños que la red neuronal da pasos más pequeños hacia los mínimos de la función de costo y finalmente deja de aprender. 
- Tanh: muy similar a la sigmoidal
- RELU: es la más utilizada para capas ocultas. Es simple de implementar y no es tan susceptible a gradientes que se desvanecen
- Softmax: no se recomienda mucho usarla ya que el cálculo de gradientes no es tan eficiente

**¿Cómo son las redes neuronales para regresión?**

$$\nu^{1} = w_{0}^{1}+w^{1}$$
$$y^{1} = \varphi(\nu^{1})$$
$$\nu^{2} = w_{0}^{2}+w^{2}y_{1}$$
<font color= #2E9AFE>$$y^{2} = \varphi(\nu^{2})$$</font>

La única diferencia es la última salida. La función de la salida tiene que ser **lineal**

**Backpropagation**

Es el método que usa para ajustar el modelo. Minimiza el error medio cuadrático (MSE), utilizando el gradiente descendente para calcular los parámetros con el orden del último hacia el primero.

**¿Cómo elegir cuántas capas ocultas?**

Se recomienda usar la regla de:
(# capas ocultas = #entradas + # salidas)

**¿Cómo elegir cuántas neuronas en las capas ocultas?**

Depende....

De forma ideal, el número de neuronas ocultas debe ser mayor al número de entradas (#Neuronas ocultas > # variables de entrada). 

Peroooo esta regla tiene varias formas de verse afectada:

1. Número de datos
2. Poder computacional

Si se tienen pocos datos, al aumentar el número de neuronas ocultas, la cantidad de pesos se empieza a ser comparable con las muestras que se tienen, entonces conviene aumentar la cantidad de neuronas ocultas. 

Si se tienen muchos datos, esta regla de (#Neuronas ocultas > # variables de entrada) no es fácil por cuestiones computacionales, por lo tanto usamos las siguientes reglas:

- El número de neuronas ocultas debe estar entre el tamaño de la capa de entrada y el tamaño de la capa de salida.
- El número de neuronas ocultas debe ser 2/3 del tamaño de la capa de entrada, más el tamaño de la capa de salida.
- El número de neuronas ocultas es la media de las neuronas en las capas de entrada y salida.

**Recomendaciones**

- Se recomienda hacer un pre-procesamiento de los datos - Escalamiento
- Comúnmente todas las capas ocultas usan la misma función de activación
- La función de salida generalmente usa diferente función de activación que el de las capas ocultas. 
- En la experiencia, conviene más aumentar el número de neuronas ocultas en lugar de aumentar el número de capas ocultas.
- Se recomienda no usar muchas capas ocultas


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

from sklearn import preprocessing
import keras
import tensorflow 
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD, Adam
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

from scikeras.wrappers import KerasClassifier, KerasRegressor
from sklearn.model_selection import cross_val_score

from sklearn.model_selection import cross_val_score

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score,mean_squared_error,make_scorer
```

#### Ejemplo de una sola variable (X) 

```python
#Generamos datos
#Generar la X
m = 1000;
X = np.linspace(-1,1,num=m)
X = np.reshape(X,(m,1))

#Generar la Y

#Y = X**2 #Función cuadrática
Y = 2*X**3+2*X**2 
#Y = 4*X*np.sin(10*X)+X*np.exp(X)*np.cos(20*X) 

[m,ne] = np.shape(X)

from sklearn import preprocessing
X = preprocessing.scale(X)
Y = preprocessing.scale(Y)

#graficando los datos
fig = plt.figure()
plt.plot(X,Y, linewidth=3)
plt.grid()
```

```python
#Dividimos los datos en test y train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size=0.2, random_state=42)
```

```python
# Model design
model = Sequential()



#Input layer with 5 neurons

#input_dim --> how many x do we have?

model.add(Dense(5, activation='relu', input_dim=1,name='input_layer'))


#If we wanted to add hidden layers
#model.add(Dense(10, activation='tanh', name='hidden_layer'))

#Output layer. Should be linear
model.add(Dense(1, activation='linear',name='output_layer'))




#SDG

#epochs = how many times does SDG repeats to find minimun MSE
#If function has a lot of spikes, its recomended to increase learning rate
# Momentum: Uses physics laws to find local optimals and makes SDG faster


#hiperparameters
learning_rate=0.1
epochs = 200
momentum = 0.8
decay_rate = learning_rate/epochs
sgd = tensorflow.keras.optimizers.legacy.SGD(learning_rate=learning_rate, momentum=momentum)


model.compile(loss='mean_squared_error',
              optimizer=sgd, 
             metrics = ['mse'])
```

```python
#Ajusting the model
history = model.fit(X_train, y_train, 
                   epochs=epochs, 
                   batch_size=100, 
                   validation_data=(X_test,y_test))
```

```python
#Plotting MSE
fig, ax = plt.subplots(1, 1, figsize=(10,6))
ax.plot(np.sqrt(history.history['mse']), 'r', label='train')
ax.plot(np.sqrt(history.history['val_mse']), 'b' ,label='val')
ax.set_xlabel(r'Epoch', fontsize=20)
ax.set_ylabel(r'MSE', fontsize=20)
ax.legend()
ax.tick_params(labelsize=20)
```

```python
#Model evaluation

#Model evaluation in test data
score = model.evaluate(X_test,y_test)
print('Test mse:', score[1])
```

```python
#Prediction and R2

#Test data
yhat = model.predict(X_test)
R2_score = r2_score(y_test,yhat)
xmin,xmax = min(y_test),max(y_test)
xline = np.linspace(xmin,xmax)
fig = plt.figure(figsize=(10,6))
plt.scatter(y_test,yhat,label='Estimation')
plt.plot(xline,xline,'k--',label='Perfect estimation')
plt.xlabel('Real output', fontsize=20)
plt.ylabel('Estimation output', fontsize=20)
plt.title('Test R^2=%0.4f'%R2_score, fontsize=20)
plt.legend()
plt.grid()
plt.show()
```

```python
print('R2:', R2_score)
```

####  Cross Validation

```python
# Function that returns a compiled neural network
def create_network():
    
    #Initializing object
    network = Sequential()

    #Adding input layer
    network.add(Dense(5, activation='tanh', input_dim=2,name='input_layer')) #Relu

    # Adding hidden layer
    #network.add(Dense(10, activation='relu')) #tanh

    #Adding linear output layer
    network.add(Dense(1, activation='linear'))
    
    #Optimizing
    opt = tensorflow.keras.optimizers.legacy.SGD(learning_rate=learning_rate,momentum=momentum)
    
    #Compile network
    network.compile(loss='mean_squared_error', 
                    optimizer=opt, 
                    metrics=['mse']) 
    
    #Return compiled network
    return network
```

```python
#Transform model to keras so it can be used by scikit-learn
neural_network = KerasRegressor(model=create_network, 
                                 epochs=200, 
                                 batch_size=100, 
                                 verbose=0)
```

```python
#Evaluate using 4-fold cross validation
print('MSE:', (-cross_val_score(neural_network, X, Y, cv=4,  scoring='neg_mean_squared_error')).mean())
```

```python
print('R2:', (cross_val_score(neural_network, X, Y, cv=4,  scoring='r2')).mean())
```

### Hiperparameter optimization

**GridSearch**

![GridSearch](https://www.yourdatateacher.com/wp-content/uploads/2021/03/image-7.png)

```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
from scikeras.wrappers import KerasRegressor

# Function to create model
def create_model():
    
    # create model
    model = Sequential()
    model.add(Dense(5, input_shape=(2,), activation='tanh'))
    model.add(Dense(1, activation='linear'))
    return model

# fix random seed for reproducibility
seed = 7
tensorflow.random.set_seed(seed)

# create model
model = KerasRegressor(model=create_model, loss='mean_squared_error',
                        optimizer=tensorflow.keras.optimizers.legacy.SGD,
                          epochs=100, batch_size=10, verbose=0)

# define the grid search parameters
learn_rate = [0.1, 0.05, 0.01]
momentum = [0.8, 0.6, 0.4]
param_grid = dict(optimizer__learning_rate=learn_rate, optimizer__momentum=momentum)


selection_score = make_scorer(r2_score) #r2
#selection_score = make_scorer(mean_squared_error,greater_is_better=False) #MSE
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=2, scoring=selection_score)
grid_result = grid.fit(X_train, y_train)

```

```python
# summarize results
print('Best: %f using %s' % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print('%f (%f) with: %r' % (mean, stdev, param))
```

```python
#Model with optimal hiperparameters
epochs = 200
learning_rate = 0.05
momentum = 0.6

# Neural net
model = Sequential()
model.add(Dense(5,activation='tanh',input_dim=2))
model.add(Dense(1,activation='linear'))

# Optimization
opt = tensorflow.keras.optimizers.legacy.SGD(learning_rate=learning_rate,momentum=momentum,
                           nesterov=True)

model.compile(loss = 'mean_squared_error',
              optimizer=opt,
              metrics=['mse'])

#Adjust model
model_history = model.fit(X_train,y_train,
                    epochs=epochs,
                    batch_size=200,
                    validation_data=(X_test,y_test))

#performance
score = model.evaluate(X_test,y_test)
print('Test loss:', score[0])
print('Test mse:', score[1])
```

```python
#Plot MSE
fig, ax = plt.subplots(1, 1, figsize=(10,6))
ax.plot(np.sqrt(model_history.history['mse']), 'r', label='train')
ax.plot(np.sqrt(model_history.history['val_mse']), 'b' ,label='val')
ax.set_xlabel(r'Epoch', fontsize=20)
ax.set_ylabel(r'MSE', fontsize=20)
ax.legend()
ax.tick_params(labelsize=20)
```

```python
#R2
yhat = model.predict(X_test)

R2_score = r2_score(y_test,yhat)

xmin,xmax = min(y_test),max(y_test)
xline = np.linspace(xmin,xmax)
fig = plt.figure(figsize=(10,6))
plt.scatter(y_test,yhat,label='Estimation')
plt.plot(xline,xline,'k--',label='Perfect estimation')
plt.xlabel('Real output', fontsize=20)
plt.ylabel('Estimation output', fontsize=20)
plt.title('Test R^2=%0.4f'%R2_score, fontsize=20)
plt.legend()
plt.grid()
plt.show()
```

### Neural networks advantages and disadvantages

#### Advantages

- If one neuron fail we can still get results
- We can work with incomplete information
- Capacity to learn by themself and generate result not limited to information given at the beggining
- Can be pararell processing
- Good for image analysis

#### Disadvantages

- Black box model
- Require more data
- Because of the data amount, they require big processor power
- Can be really complex
- No formula to identify hiperparameters and best options
